# Make a new dataframe that contains only those songs that actually have lyrics_features. 

```python
df_songs_lyrics =  songs_dataset[songs_dataset['lyrics_features'].str.len()!=0]
df_songs_lyrics.shape
```

Reset indext 

```python
df_songs_lyrics.reset_index(drop=True, inplace=True)
```

```python
df_songs_lyrics
```

Next,build a list/array containing the lyrics_features that you are going to be using for your NLP processing. 

```python
lyrics_list = []
lyrics_column = df_songs_lyrics['lyrics_features']
for lyrics in lyrics_column:
    lyrics_list.append(' '.join(lyrics))

#checked
print(len(lyrics_list))
```

```python
# Clean word function from lab3
import string
from string import punctuation
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS
from nltk.stem.snowball import SnowballStemmer

translator = str.maketrans('', '', punctuation)
stemmer = SnowballStemmer('english')

def clean_text(raw_text):
    clean_words = []
    raw_text.lower()
    raw_text = raw_text.translate(translator)
    split_words = raw_text.split()
    
    for word in split_words:
        if word not in ENGLISH_STOP_WORDS:
            word = stemmer.stem(word)
            clean_words.append(word)
            
    return ' '.join(clean_words)
```

```python
lyrics_list_cleaned = []
for lyrics in lyrics_list:
    lyrics_list_cleaned.append(clean_text(lyrics))
```

Add column lyrics_features_clean to df_songs_lyrics

```python
df_songs_lyrics['lyrics_features_clean'] = lyrics_list_cleaned
df_songs_lyrics
```

```python
# df_songs_lyrics.shape
df_songs_lyrics['moods'].shape
```

Build your bag of words and tf-idf features using the lyrics_features from only the celebratory and sad songs. Process  in moods columns

```python
mood_filter = []
for mood in df_songs_lyrics['moods']:
    if 'celebratory' in mood:
        mood_filter.append('celebratory')
    elif 'sad' in mood:
        mood_filter.append('sad')
    else:
        mood_filter.append(mood)
        
len(mood_filter)
# add to df_songs_lyrics song
# df_songs_lyrics['mood_filter'] = mood_filter
```

```python
df_songs_lyrics['mood_filter'] = mood_filter
df_songs_lyrics
```

Get celebratory song only

```python
songs_celebratory = df_songs_lyrics.loc[(df_songs_lyrics['mood_filter']=='celebratory')]
songs_celebratory.shape
```

Get sad song only

```python
songs_sad = df_songs_lyrics.loc[(df_songs_lyrics['mood_filter']=='sad')]
songs_sad.shape
```

```python
songs_celebratory.sample(n=1249, replace=True)
songs_celebratory.shape
```

```python
# get final data
df_songs_final =  pd.concat([songs_celebratory, songs_sad], axis=0)
df_songs_final.shape
```

# Bag Of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
bag_of_words = count_vect.fit_transform(df_songs_final['lyrics_features_clean'])
bag_of_words
print(bag_of_words)
```

```python
bag_of_words.shape
```

## Bag of Words - Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report

X = bag_of_words
y = df_songs_final['lyrics_features_clean']

model_loreg = LogisticRegression(solver='saga',multi_class='multinomial')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model_loreg.fit(X_train, y_train)
model_loreg_predictions = model_loreg.predict(X_test)

print("#BOW LogisticRegression")
print(accuracy_score(y_test, model_loreg_predictions))
```

## Bag of Words - Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

X = bag_of_words
y = df_songs_final['lyrics_features_clean']

model_rf = RandomForestClassifier(n_estimators=5,min_samples_split=2, max_features='log2')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model_rf.fit(X_train, y_train)
model_rf_predictions = model_rf.predict(X_test)

print("#BOW Random Forest")
print(accuracy_score(y_test, model_rf_predictions))
```

## Bag of Words - Support Vector Machines (SVM)

```python
from sklearn.svm import SVC

svc = SVC(C=1, gamma=1)
X = bag_of_words
y = df_songs_final['lyrics_features_clean']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
svc.fit(X_train, y_train)
svc_bow_predictions = svc.predict(X_test)

print("#BOW Support Vector Machines")
print(accuracy_score(y_test, svc_bow_predictions))
```

```python
X = tf_idf
y = df_songs_final['lyrics_features_clean']

model_rf = RandomForestClassifier(n_estimators=5,min_samples_split=2, max_features='log2')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model_rf.fit(X_train, y_train)
model_rf_predictions = model_rf.predict(X_test)

print("#TF-IDF Random Forest")
print(accuracy_score(y_test, model_rf_predictions))
```
